```python
train_data = pd.read_csv('/kaggle/input/spaceship-titanic/train.csv', index_col='PassengerId')
train_data.head()
```

```python
train_data.shape
```

```python
del train_data['Name']
train_data.info()
```

```python
train_data.isna().sum()
```

```python
train_data = train_data.fillna(train_data.mean())
train_data.isna().sum()
```

```python
train_data['Cabin'] = train_data['Cabin'].fillna(train_data['Cabin'].mode()[0])
train_data['Destination'] = train_data['Destination'].fillna(train_data['Destination'].mode()[0])
train_data['HomePlanet'] = train_data['HomePlanet'].fillna(train_data['HomePlanet'].mode()[0])
train_data.isna().sum()
```

```python
train_data.head()
```

```python
del train_data['ShoppingMall']
del train_data['FoodCourt']
del train_data['Cabin']
del train_data['HomePlanet']
del train_data['Destination']
train_data.shape
```

```python
train_data = pd.get_dummies(train_data)
train_data.shape
```

```python
train_data.head()
```

```python
del train_data['CryoSleep_0.35830580462482303']
del train_data['VIP_0.023439340400471142']
train_data.head()
```

```python
x = train_data.loc[:, train_data.columns != 'Transported']
x.head()
```

```python
y = train_data['Transported']
y.head()
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(x,y)
```

```python
from sklearn.metrics import precision_score,recall_score
y_pred = clf.predict(x)
precision_score(y,y_pred)
```

```python
test_data = pd.read_csv('/kaggle/input/spaceship-titanic/test.csv', index_col='PassengerId')
del test_data['Name']
del test_data['ShoppingMall']
del test_data['FoodCourt']
del test_data['Cabin']
del test_data['HomePlanet']
del test_data['Destination']
test_data.head()
```

```python
test_data = pd.get_dummies(test_data)
test_data.head()
```

```python
test_data.isnull().sum()
```

```python
test_data = test_data.fillna(test_data.mean())
test_data.isna().sum()
```

```python
predict = clf.predict(test_data)
```

```python
output = pd.DataFrame(columns=['PassengerId', 'Transported'])
output['PassengerId'] = test_data.index
output['Transported'] = predict
output.to_csv('submission.csv', index=False)
print('Submission succesful!')
```
